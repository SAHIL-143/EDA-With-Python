```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# Upload the Excel file
uploaded = files.upload()

# Replace with the correct filename after upload
file_name = list(uploaded.keys())[0]

# Load the Excel file and parse the sheet
df = pd.read_excel(file_name, sheet_name="EXERCISE 7.4")

# Inspect to identify the correct data range and number of columns
preview = df.iloc[0:20]
print(preview)

# Extract and clean the relevant part of the data (columns D, E, F -> index 3,4,5)
data = df.iloc[7:19, 3:6].copy()
data.columns = ["Aspect", "Our store", "All stores"]
data.reset_index(drop=True, inplace=True)

# Convert values to numeric
data["Our store"] = pd.to_numeric(data["Our store"], errors='coerce')
data["All stores"] = pd.to_numeric(data["All stores"], errors='coerce')

# Set style
sns.set(style="whitegrid")

# Plotting
plt.figure(figsize=(12, 8))
bar_width = 0.4
index = range(len(data))

# Bar chart
our_store_bars = plt.barh([i + bar_width for i in index], data['Our store'], height=bar_width, label='Our store', color='skyblue')
all_stores_bars = plt.barh(index, data['All stores'], height=bar_width, label='All stores', color='lightgreen')

# Add data labels to each bar
for bar in our_store_bars:
    plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2, f'{bar.get_width():.1f}%', va='center')

for bar in all_stores_bars:
    plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2, f'{bar.get_width():.1f}%', va='center')

# Y-axis labels
plt.yticks([i + bar_width / 2 for i in index], data['Aspect'])
plt.xlabel('% Favorable')
plt.title('Customer Survey: Our Store vs. All Stores')
plt.legend()
plt.tight_layout()
plt.show()
```
