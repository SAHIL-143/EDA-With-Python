# **What is the average exam score, and how is the performance distributed across students?**

```python
Average_Exam_Score = df['Exam_Score'].mean()
print(f'Average_Exam_Score: {Average_Exam_Score}')
```

# **How does distance from home to school affect attendance rates and ultimately academic performance?**

```python

#Step 1: Convert 'Distance_from_Home' to a numerical scale

distance_mapping = {'Near': 1, 'Moderate': 2, 'Far': 3}   # Use of dictanory for mapping values
df['Distance_Numerical'] = df['Distance_from_Home'].map(distance_mapping) #Map values of Series according to an input mapping or function

# Step 2: Correlation between Distance from Home and Attendance

distance_attendance_corr = df['Distance_Numerical'].corr(df['Attendance']) #corr() is used to find the pairwise correlation of all columns
print(f"Correlation between Distance from Home and Attendance: {distance_attendance_corr:.4f}")

# Step 3: Correlation between Attendance and Exam Scores (Academic Performance)

attendance_performance_corr = df['Attendance'].corr(df['Exam_Score'])
print(f"Correlation between Attendance and Exam Scores: {attendance_performance_corr:.4f}")

# Step 4: Visualize the relationship between Distance and Attendance
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Distance_Numerical', y='Attendance', data=df)
plt.title('Distance from Home vs Attendance')
plt.xlabel('Distance from Home (1 = Near, 2 = Moderate, 3 = Far)')
plt.ylabel('Attendance (%)')
plt.grid(True)
plt.show()

# Step 5: Visualize the relationship between Attendance and Exam Scores
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Attendance', y='Exam_Score', data=df)
plt.title('Attendance vs Exam Scores')
plt.xlabel('Attendance (%)')
plt.ylabel('Exam Scores')
plt.grid(True)
plt.show()

```

**Is there a relationship between sleeping hours and performance on exams?**

```python
sleep_correlation = df['Sleep_Hours'].corr(df['Exam_Score'])  #df['Sleep_Hours'].corr(df['Exam_Score']) to find correlation between the number of sleep hours and exam scores.
print(f"Correlation between Sleep Hours and Exam Score: {sleep_correlation:.4f}")


# ploting sleep hours and exam score
plt.figure(figsize=(8, 6))
sns.regplot(x='Sleep_Hours', y='Exam_Score', data=df, scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.title('Relationship between Sleep Hours and Exam Performance')
plt.xlabel('Sleep Hours')
plt.ylabel('Exam Score')
plt.grid(True)
plt.show()
```

Analysis:
A negative correlation which is -0.0170 implies that Less sleep is associated with better performance.

What impact does parental involvement have on exam scores for students in different income brackets?

```python
# Group by Family Income and Parental Involvement
grouped = df.groupby(['Family_Income', 'Parental_Involvement'])['Exam_Score'].mean().reset_index()   #groupby function calculate the average exam scores for family income and parental involvement.

# Pivot the DataFrame for better visualization
pivot_table = grouped.pivot(index='Family_Income', columns='Parental_Involvement', values='Exam_Score')
#sleep_correlation = df['Sleep_Hours'].corr(df['Exam_Score'])  #df['Sleep_Hours'].corr(df['Exam_Score']) to find correlation between the number of sleep hours and exam scores.
print(f"Correlation between Parental_Involvement and Exam Score: {sleep_correlation:.4f}")
# Plotting the results
pivot_table.plot(kind='bar', figsize=(10, 6))
plt.title('Average Exam Scores by Family Income and Parental Involvement')
plt.xlabel('Family Income')
plt.ylabel('Average Exam Score')
plt.xticks(rotation=0)
plt.legend(title='Parental Involvement')
plt.grid(axis='y')
plt.show()
```

# **Correlation between parental involvement and exam score is -ve 0.0170 there is little to no linear relationship between parental involvement and exam scores**

```python
grouped = df.groupby('Parental_Education_Level').agg(
    Average_Exam_Score=('Exam_Score', 'mean'),
    Resource_Access=('Access_to_Resources', lambda x: x.mode()[0])
).reset_index()

# Plotting the average exam scores
plt.figure(figsize=(10, 6))
sns.barplot(data=grouped, x='Parental_Education_Level', y='Average_Exam_Score')
plt.title('Average Exam Scores by Parental Education Level')
plt.xlabel('Parental Education Level')
plt.ylabel('Average Exam Score')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# Display the grouped data
print(grouped)
```

```python
# Group by School Type and Internet Access, calculating average exam scores
grouped = df.groupby(['School_Type', 'Internet_Access'])['Exam_Score'].mean().reset_index()

# Plotting the results
plt.figure(figsize=(10, 6))
sns.barplot(data=grouped, x='School_Type', y='Exam_Score', hue='Internet_Access')
plt.title('Average Exam Scores by Internet Access in Urban vs. Rural Areas')
plt.xlabel('School Type')
plt.ylabel('Average Exam Score')
plt.legend(title='Internet Access')
plt.grid(axis='y')
plt.show()

print(grouped)
```
